```python
import pandas as pd
```

```python
df = pd.read_csv('/Volumes/Data2/RST/notebook/networks/AM-TE-16-5-constrained.csv')
```

```python
df_late = df[df['hr_o'].isin([9,10,11])]
```

```python
df = df[df['hr_o'].isin([7,8])]
```

```python
df = df[['INT_ID_o', 'INT_ID_d', 'cost']]
df_late = df_late[['INT_ID_o', 'INT_ID_d', 'cost']]

```

```python
df_tt = df.groupby(['INT_ID_o', 'INT_ID_d']).std().reset_index().rename(columns = {'cost':'std'}).merge(
    df.groupby(['INT_ID_o', 'INT_ID_d']).mean().reset_index())

df_tt_late = df_late.groupby(['INT_ID_o', 'INT_ID_d']).std().reset_index().rename(columns = {'cost':'std'}).merge(
    df_late.groupby(['INT_ID_o', 'INT_ID_d']).mean().reset_index())
```

```python
df_tt['std'] = df_tt['std'].fillna(0)
df_tt_late['std'] = df_tt_late['std'].fillna(0)
```

```python
for period in ['EM','AM','MD','PM','EV']:
    
    df = pd.read_csv('/Volumes/Data2/RST/notebook/networks/' + period +'-TE-16-8.csv')
    
    if period == 'EV':
        hr_list = [20,21]
        hr_list_late = [22,23,0]
    else:
        hr_list = list(df['hr_o'].sort_values().drop_duplicates())[0:2]
        hr_list_late = list(df['hr_o'].sort_values().drop_duplicates())[2:5]
        
    df = df[df['hr_o'].isin(hr_list)]
    
    df = df[df['hr_o'].isin(hr_list_late)]


    df = df[['INT_ID_o', 'INT_ID_d', 'cost']]
    df_late = df_late[['INT_ID_o', 'INT_ID_d', 'cost']]
    

    
    
    df_tt_late = df_late.groupby(['INT_ID_o', 'INT_ID_d']).std().reset_index().rename(columns = {'cost':'std'}).merge(
        df_late.groupby(['INT_ID_o', 'INT_ID_d']).mean().reset_index())
    
    df_tt = df.groupby(['INT_ID_o', 'INT_ID_d']).std().reset_index().rename(columns = {'cost':'std'}).merge(
        df.groupby(['INT_ID_o', 'INT_ID_d']).mean().reset_index())
    df_tt['std'] = df_tt['std'].fillna(0)
    df_tt['std'] = df_tt['std'].fillna(0)


        
    df_tt_excluded = df_tt[['INT_ID_o', 'INT_ID_d']].copy()
    df_tt_excluded['early'] = True
    df_tt_excluded = df_tt_excluded.merge(df_tt_late, how = 'outer')
    df_tt_excluded = df_tt_excluded[df_tt_excluded['early']!=True].copy()
    
    df_tt['std'] = df_tt['std'].fillna(0)

    df_tt = df_tt.append(df_tt_excluded[['INT_ID_o', 'INT_ID_d', 'std', 'cost']])


    
    df_tt.to_csv(period + '_segment-cost.csv', index = False)
```
