```python
import geopandas as gpd
import pandas as pd
```

```python
simplify_dir = 'GTFS/stop_snapping/GIS Simplify/'
gtfs_dir = 'GTFS/TTC_2016-10-03/'
```

# Cleaning up the TTS boundary file

```python
tts = gpd.read_file('GIS/tts2016_2006zn_shp/tts2016_2006zn_2018_region.shp')
```

```python
tts = tts.to_crs(epsg='32617')
```

```python
tts = tts[tts['region']==1]
```

```python
tts
```

```python
cda = gpd.read_file('GIS/cda/lda_000b16a_e.shp')
```

```python
population = pd.read_csv('GIS/census_population.csv')
```

```python
cda = cda[cda['PRNAME'] == 'Ontario']
cda = cda[cda['CDNAME'] == 'Toronto']
cda
```

```python
cda = cda[['DAUID', 'geometry']]
cda['DAUID'] = cda['DAUID'].astype(int)
```

```python
cda = cda.to_crs(epsg = '32617')
```

```python
da_list = list(cda['DAUID'].astype(int).drop_duplicates())
```

```python
population = population[(population['GEO_CODE (POR)'].astype(str).str[:4] == '3520') & (population['GEO_LEVEL'] == 4)]
population = population[population['DIM: Age (in single years) and average age (127)'] == 'Total - Age']
population = population[population['GEO_CODE (POR)'].isin(da_list)]
population = population.reset_index()
```

```python
population = population[['GEO_CODE (POR)', 'Dim: Sex (3): Member ID: [1]: Total - Sex']]
```

```python
population = population.rename(columns = {'GEO_CODE (POR)':'DAUID', 'Dim: Sex (3): Member ID: [1]: Total - Sex':'population'})
```

```python
cda = cda.merge(population, how = 'left')
cda
```

```python
cda['population'].hist()
```

```python
cda['population'] = cda['population'].fillna(0)
```

```python
cda.plot()
```

```python
cda['area'] = cda.area
```

```python
cda.to_file('GIS/cda_pop.geojson', driver = 'GeoJSON')
```

```python
from geopandas.tools import overlay
cda_split = overlay(cda, tts, how="intersection")
```

```python
cda_split.plot()
```

```python
cda_split = cda_split.rename(columns = {'area':'area_original'})
cda_split['area_split'] = cda_split.area
```

```python
cda_split['pop_split'] = (cda_split['area_split']/cda_split['area_original'])*cda_split['population']
```

```python
cda_split['split_centroid'] = cda_split.centroid
```

```python
cda_split['split_centroid_x'] = cda_split.centroid.x
cda_split['split_centroid_y'] = cda_split.centroid.y
```

```python
cda_split
```

```python
cda_split[['DAUID', 'gta06', 'area_original', 'area_split']].to_csv('GIS/cda_split_tts.csv', index = False)
```

```python
table = []
for zone in list(tts['gta06']):
    temp = cda_split[cda_split['gta06'] == zone].copy()
    temp['x_weight'] = temp['pop_split'] * temp['split_centroid_x']
    x_sum = temp['x_weight'].sum()
    pop_sum = temp['pop_split'].sum()
    
    temp['y_weight'] = temp['pop_split'] * temp['split_centroid_y']
    y_sum = temp['y_weight'].sum()  
    
    table.append([zone, x_sum/pop_sum, y_sum/pop_sum, pop_sum])
    
```

```python
centroids = pd.DataFrame.from_records(table, columns = ['gta06', 'x', 'y', 'pop'])
```

```python
centroids['pop'].hist()
```

```python
gdf = gpd.GeoDataFrame(
    centroids, geometry=gpd.points_from_xy(centroids.x, centroids.y))
```

```python
tts_centroids = gdf.dropna().set_crs(epsg = '32617').to_crs(epsg = '4326')[['gta06', 'geometry']]
```

```python
tts_centroids['lat'] = tts_centroids.geometry.y
tts_centroids['lon'] = tts_centroids.geometry.x
```

```python
tts_centroids
```

```python
tts['gta06'] = tts['gta06'].astype(int)
```

```python
tts = tts.merge(tts_centroids[['gta06','lat','lon']], how = 'left')
```

```python
import numpy as np
```

```python
tts['lat'] = np.where(tts['lat'].isna(), tts.to_crs(epsg = '4326').centroid.y, tts['lat'])
tts['lon'] = np.where(tts['lon'].isna(), tts.to_crs(epsg = '4326').centroid.x, tts['lon'])
```

```python
tts = tts[['gta06', 'pd', 'geometry', 'lat', 'lon']]
```

```python
tts.to_file('GIS/tts.geojson', driver = 'GeoJSON')
```

```python
int_stop = pd.read_csv(simplify_dir + 'int_stop.csv')
```

```python
int_stop = gpd.GeoDataFrame(
    int_stop, geometry=gpd.points_from_xy(int_stop.int_lon, int_stop.int_lat))
```

```python
int_stop = int_stop.set_crs(epsg = '4326')
int_stop_buffer = int_stop.copy().to_crs(epsg = '32617')
int_stop_buffer['geometry'] = int_stop_buffer.buffer(50)
int_tts = gpd.sjoin(tts.to_crs(epsg = '32617'), int_stop_buffer, op = 'intersects', how = 'right')[['gta06', 'INT_ID']]
```

```python
int_tts
```

```python
int_tts.to_csv('GIS/int_tts.csv', index = False)
```
