```python
import pandas as pd
import re
import os
```

```python
tts_od = pd.read_csv('LICO_TTS.csv')
```

```python
tts_od = tts_od.groupby(['origin', 'destination', 'type']).sum().reset_index()
tts_od = tts_od[['origin', 'destination','total', 'type']]
```

```python
file = open('raw-tts_c0.txt', 'r').read().split('\n')[25:]
file
```

```python
table = []
for row in file:

    if row[0:5] == 'TABLE':
        period_int = int(re.split('[()]', row)[-2])
        if period_int == 2:
            period = 'EM'
        elif period_int == 4:
            period = 'AM'
        elif period_int == 6:
            period = 'MD'
        elif period_int == 8:
            period = 'PM'
        elif period_int == 10:
            period = 'EV'
        else:
            pass
        continue
    elif row == '':

        continue
    elif row == '  gta06_orig  gta06_dest      total':

        continue
    else:
        parsed_row = row.split()
        parsed_row.append(period)
        table.append(parsed_row)
```

```python
c0_df = pd.DataFrame.from_records(table, columns = ['origin', 'destination', 'total', 'period'])
```

```python
c0_df['type'] = 'c0_' + c0_df['period'].str.lower()
c0_df = c0_df[['origin', 'destination','total', 'type']]
tts_od = tts_od.append(c0_df)
```

```python
tts_od.to_csv('LICO_TTS.csv', index = False)
```
