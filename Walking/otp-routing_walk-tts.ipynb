```python
start_time = '8:00'
date = '2019-12-01'
mode = 'WALK'
searchWindow = 10
threads = 64
start = 0
```

```python
def return_walk(param):
    ox = param[0]
    oy = param[1]
    dx = param[2]
    dy = param[3]
    options = {
        'fromPlace': str(oy) + ", " + str(ox),
        'toPlace': str(dy) + ", " + str(dx),
        'time': start_time,
        'date': date,
        'mode':mode,
        'maxWalkDistance':10000,
        'clampInitialWait':0,
        'wheelchair':False,
        'searchWindow':searchWindow * 60
        #'bannedStops':banned
        #'bannedRoutes':banned_routes
    }

    response = requests.get(
        "http://localhost:8080/otp/routers/default/plan",
        params = options
           )
    # return as json
    data = json.loads(response.text)
    return data
```

```python
tts_difference = tts_int.merge(stops_old[['INT_ID', 'flag']], how = 'left').copy()
tts_difference = tts_difference[tts_difference['flag'].isna()].reset_index().drop(columns = ['index'])
tts_difference
```

```python
tts_int
```

```python
start = 0
table = []
for index, row in tts_int.groupby(tts_int.index // threads):
    if index % 100 == 0:
        print(index * threads)
    param = []
    dest = []
    origin = []
    thread_lst = []
    result = []
    
    for j in range(start,start+threads):
                
        try:
            tts = row['gta06'][j]
            x1 = row['lon'][j]
            y1 = row['lat'][j]
            int_id = int(row['INT_ID'][j])
            x2 = row['int_lon'][j]
            y2 = row['int_lat'][j]
        except:
            x2 = 0
            y2 = 0
            tts = 0
            x1 = 0
            y1 = 0
            int_id = 0
        param.append((x1,y1,x2,y2))
        dest.append(int_id)
        origin.append(tts)

        
    
    with concurrent.futures.ThreadPoolExecutor() as executor:
        thread_lst = executor.map(return_walk, param)
        result = list(thread_lst)
    for c in range(0, threads):
        try:
            table.append([origin[c], dest[c], result[c]['plan']['itineraries'][0]['duration']])
        except:
            continue
        
        
    start = start + threads
```

```python
walk_times = pd.DataFrame.from_records(table, columns = ['gta06', 'INT_ID', 'duration'])
```
