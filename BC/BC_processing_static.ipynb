```python
import pandas as pd
import os
```

```python
save_dir = '/Volumes/Data2/RST/notebook/networks/'
```

```python
period = 'AM'
df_list = []
for filename in os.listdir(period):
    try:
        if filename.split('_')[1] == 'edge-raw.csv':
            df_list.append(pd.read_csv(period + '/' + filename))
        else:
            continue
    except:
        continue
```

```python
am_edge = pd.concat(df_list)
am_edge['o_node'] = am_edge['o_node_time'].str.split('-', expand = True)[0]
am_edge['d_node'] = am_edge['d_node_time'].str.split('-', expand = True)[0]
am_edge.to_csv(save_dir + period + '/BC-' + period + '_edge-raw.csv', index = False)
```

```python
am_edge
```

```python
edge_master = am_edge[['INT_ID_o', 'INT_ID_d', 'o_node', 'd_node', 'route']].drop_duplicates()
```

```python
time_master = am_edge[['o_node_time', 'd_node_time', 'o_node', 'd_node']].drop_duplicates()
```

```python
key = time_master.merge(edge_master, how = 'left')
```

```python
am_edge = key.merge(am_edge, how = 'left')[['o_node', 'd_node', 'route', 'INT_ID_o', 'INT_ID_d', 'bc_single']]
am_edge['bc_single'] = am_edge['bc_single'].fillna(0)
am_edge.to_csv(save_dir + period + '/BC-' + period + '_edge.csv', index = False)
am_edge
```

```python
am_edge = am_edge[['o_node', 'd_node', 'INT_ID_o', 'INT_ID_d', 'route', 'bc_single']]
```

```python
am_edge_avg = am_edge.groupby(['o_node', 'd_node', 'INT_ID_o', 'INT_ID_d', 'route']).mean().reset_index()
```

```python
am_edge_avg
```

```python
am_edge_avg.to_csv(save_dir + period + '/BC-' + period + '_edge-avg.csv', index = False)
```

```python
period_list = ['EM','PM','MD', 'EV']

for period in period_list:

    df_list = []
    for filename in os.listdir(period):
        try:
            if filename.split('_')[1] == 'edge-raw.csv':
                df_list.append(pd.read_csv(period + '/' + filename))
            else:
                continue
        except:
            continue
    edge = pd.concat(df_list)
    edge['o_node'] = edge['o_node_time'].str.split('-', expand = True)[0]
    edge['d_node'] = edge['d_node_time'].str.split('-', expand = True)[0]
    edge.to_csv(save_dir + period + '/BC-' + period + '_edge-raw.csv', index = False)

    edge_master = edge[['INT_ID_o', 'INT_ID_d', 'o_node', 'd_node', 'route']].drop_duplicates()

    time_master = edge[['o_node_time', 'd_node_time', 'o_node', 'd_node']].drop_duplicates()

    key = time_master.merge(edge_master, how = 'left')

    edge = key.merge(edge, how = 'left')[['o_node', 'd_node', 'INT_ID_o', 'INT_ID_d', 'route','bc_single']]
    edge['bc_single'] = edge['bc_single'].fillna(0)
    edge.to_csv(save_dir + period + '/BC-' + period + '_edge.csv', index = False)
    edge_avg = edge.groupby(['o_node', 'd_node', 'INT_ID_o', 'INT_ID_d', 'route']).mean().reset_index()
    edge_avg.to_csv(save_dir + period + '/BC-' + period + '_edge-avg.csv', index = False)
```

```python
period_list = ['EM','PM','MD', 'EV']

for period in period_list:
    
    df_list = []
    for filename in os.listdir(period):
        try:
            if filename.split('_')[1] == 'raw.csv':
                df_list.append(pd.read_csv(period + '/' + filename))
            else:
                continue
        except:
            continue

    node = pd.concat(df_list)
    node.to_csv(save_dir + period + '/BC-' + period + '_node-raw.csv', index = False)

    node_master = node[['INT_ID', 'o_node', 'd_node']].drop_duplicates()

    key_time = node[['o_node_time', 'd_node_time', 'o_node', 'd_node']].drop_duplicates()
    node_filled = key_time.merge(node_master, how = 'left').merge(node, how = 'left')[['o_node_time', 'd_node_time', 'o_node', 'd_node', 'INT_ID', 'bc_single']]

    node_filled['bc_single'] = node_filled['bc_single'].fillna(0)
    node_filled.to_csv(save_dir + period + '/BC-' + period + '_node.csv', index = False)
    
    node_filled = node_filled[['o_node', 'd_node', 'INT_ID', 'bc_single']]
    node_avg = node_filled.groupby(['o_node', 'd_node', 'INT_ID']).mean().reset_index()
    node_avg.to_csv(save_dir + period + '/BC-' + period + '_node-avg.csv', index = False)
```
