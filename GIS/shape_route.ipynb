```python
import geopandas as gpd
import pandas as pd
```

```python
shapes = gpd.read_file('/Volumes/Data2/RST/notebook/GTFS/TTC_2016-10-03/shapes.geojson')
shapes
```

```python
trips = pd.read_csv('/Volumes/Data2/RST/notebook/GTFS/TTC_2016-10-03/trips.txt')
trips
```

```python
routes = pd.read_csv('/Volumes/Data2/RST/notebook/GTFS/TTC_2016-10-03/routes.txt')
routes
```

```python
shape_route = shapes.merge(trips, how = 'left').merge(routes)[[
    'route_short_name', 'trip_headsign','geometry']].drop_duplicates()
```

```python
shape_route
```

```python
for index, row in shape_route.iterrows():
    
    if int(row['route_short_name']) < 4:
        continue
    
    express = row['trip_headsign'].split(' ')[2]
    
    if express[-1] != 'E':
        if express != '53F':
            if express != '60F':#specific exception for steeles
                continue
    
    shape_route.at[index, 'route_short_name'] = str(int(row['route_short_name']) + 9000)
```

```python
shape_route['route_short_name'] = shape_route['route_short_name'].astype(int)
shape_route = shape_route.sort_values(by = 'route_short_name')
shape_route
```

```python
shape_route = shape_route.drop_duplicates(subset = ['route_short_name', 'geometry'])
```

```python
shape_route.to_file('/Volumes/Data2/RST/notebook/GTFS/TTC_2016-10-03/shapes_route.geojson', driver = 'GeoJSON')
```
