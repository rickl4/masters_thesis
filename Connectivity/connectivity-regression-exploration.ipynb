```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
import scipy.stats as stats
from matplotlib.offsetbox import AnchoredText
plt.rcParams['figure.dpi'] = 450
```

```python
font = {'size'   : 8}

mpl.rc('font', **font)
```

```python
directory = '/Volumes/Data2/RST/notebook/'
```

```python
ward = pd.read_csv(directory + 'Ward Attributes/ward_profiles.csv')
```

```python
ward
```

```python
period = 'AM'

conn = pd.read_csv(directory + 'networks/' + period + '/' + period + '_connectivity.csv')
conn = conn.merge(ward)

fig, ax = plt.subplots(3,2)

fig.set_size_inches(6,10)

i = 0
j = 0

for group in ['bl', 'imm', 'lep', 'rac', 'lico', 'c0']:

    x = conn[group]
    y = conn['connectivity']

    r, p = stats.pearsonr(x, y)

    sns.regplot(data = conn, x = x, y = y, ax = ax[i,j], scatter_kws={'s':6})
    anchored_text = AnchoredText('r = ' + str(round(r,3)) +', p = ' + str(round(p,3)) , loc=2, 
                                prop=dict(fontsize = 8))

    ax[i,j].add_artist(anchored_text)

    i = i + 1

    if i > 2:
        i = 0
        j = j + 1
plt.suptitle(period + ' Period')

plt.tight_layout()

plt.savefig('connectivity-reg_' + period + '.png')
```
