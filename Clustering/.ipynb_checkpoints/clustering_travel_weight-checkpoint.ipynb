```python
dest_con = ward_trips.merge(connectivity[['Ward', 'connectivity', 'period']], left_on = ['ward_d', 'period'], right_on = ['Ward', 'period'])
dest_con['weighted_con'] = dest_con['connectivity'] * dest_con['total']  
dest_con = dest_con.groupby(['group', 'period']).sum()[['weighted_con']].reset_index().sort_values(
    by = ['period', 'weighted_con'], ascending =  False)

dest_con = dest_con.merge(ward_trip_total)
dest_con['avg_con'] = dest_con['weighted_con']/dest_con['total']
dest_con.sort_values(by = ['period', 'avg_con'], ascending =  False)
```

```python
od_con = ward_trips.merge(connectivity[['Ward', 'connectivity', 'period']], left_on = ['ward_o', 'period'], right_on = ['Ward', 'period'])
od_con = od_con.merge(connectivity[['Ward', 'connectivity', 'period']], left_on = ['ward_d', 'period'], right_on = ['Ward', 'period'])
od_con['od_con'] = (od_con['connectivity_x'] + od_con['connectivity_y'])/2
```

```python
od_con['weighted_con'] = od_con['od_con'] * od_con['total']  
od_con = od_con.groupby(['group', 'period']).sum()[['weighted_con']].reset_index().sort_values(
    by = ['period', 'weighted_con'], ascending =  False)

od_con = od_con.merge(ward_trip_total)
od_con['avg_con'] = od_con['weighted_con']/od_con['total']
od_con.sort_values(by = ['period', 'avg_con'], ascending =  False)
```
