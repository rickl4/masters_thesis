```python
import pandas as pd
```

```python
cda_split = pd.read_csv('GIS/cda_split_tts.csv')
cda_pop = pd.read_csv('TTS OD/cda_pop.csv')
cda_25 = pd.read_csv('TTS OD/cda_25.csv')
```

```python
cda_pop = cda_split.merge(cda_pop)
cda_pop['male_pr'] = cda_pop['area_split']*cda_pop['male']/cda_pop['area_original']
cda_pop['female_pr'] = cda_pop['area_split']*cda_pop['female']/cda_pop['area_original']

cda_pop['male_lep_pr'] = cda_pop['area_split']*cda_pop['male_lep']/cda_pop['area_original']
cda_pop['female_lep_pr'] = cda_pop['area_split']*cda_pop['female_lep']/cda_pop['area_original']
```

```python
cda_pop = cda_pop.groupby('gta06').sum()[['male_pr', 'female_pr', 'male_lep_pr', 'female_lep_pr']].reset_index()
```

```python
cda_pop['lep_m'] = cda_pop['male_lep_pr']/cda_pop['male_pr']
cda_pop['lep_f'] = cda_pop['female_lep_pr']/cda_pop['female_pr']
cda_pop['hh'] = cda_pop['gta06']
cda_pop = cda_pop[['hh', 'lep_m', 'lep_f']]
```

```python
cda_frac = pd.wide_to_long(cda_pop, stubnames=['lep'], i='hh', j='gender', sep = '_', suffix=r'\w+').reset_index()
```

```python
cda_25 = cda_split.merge(cda_25)
cda_25['male_pr'] = cda_25['area_split']*cda_25['male']/cda_25['area_original']
cda_25['female_pr'] = cda_25['area_split']*cda_25['female']/cda_25['area_original']

cda_25['male_rac_pr'] = cda_25['area_split']*cda_25['male_rac']/cda_25['area_original']
cda_25['female_rac_pr'] = cda_25['area_split']*cda_25['female_rac']/cda_25['area_original']
```

```python
cda_25['male_bl_pr'] = cda_25['area_split']*cda_25['male_bl']/cda_25['area_original']
cda_25['female_bl_pr'] = cda_25['area_split']*cda_25['female_bl']/cda_25['area_original']
```

```python
cda_25['male_ab_pr'] = cda_25['area_split']*cda_25['male_ab']/cda_25['area_original']
cda_25['female_ab_pr'] = cda_25['area_split']*cda_25['female_ab']/cda_25['area_original']
```

```python
cda_25['male_imm_pr'] = cda_25['area_split']*cda_25['male_imm']/cda_25['area_original']
cda_25['female_imm_pr'] = cda_25['area_split']*cda_25['female_imm']/cda_25['area_original']
```

```python
cda_25 = cda_25.groupby('gta06').sum()[['male_pr', 'female_pr', 'male_bl_pr', 'female_bl_pr',
                                        'male_ab_pr', 'female_ab_pr',
                                        'male_imm_pr', 'female_imm_pr',
                                        'male_rac_pr', 'female_rac_pr']].reset_index()
```

```python
cda_25['ab_m'] = cda_25['male_ab_pr']/cda_25['male_pr']
cda_25['ab_f'] = cda_25['female_ab_pr']/cda_25['female_pr']
```

```python
cda_25['rac_m'] = cda_25['male_rac_pr']/cda_25['male_pr']
cda_25['rac_f'] = cda_25['female_rac_pr']/cda_25['female_pr']
```

```python
cda_25['imm_m'] = cda_25['male_imm_pr']/cda_25['male_pr']
cda_25['imm_f'] = cda_25['female_imm_pr']/cda_25['female_pr']
```

```python
cda_25['hh'] = cda_25['gta06']
cda_25 = cda_25[['hh', 'rac_m', 'rac_f', 'bl_f', 'bl_m', 'ab_m', 'ab_f', 'imm_m', 'imm_f']]
```

```python
cda_frac = cda_frac.merge(pd.wide_to_long(cda_25, stubnames=['rac', 'bl', 'ab', 'imm'], 
                                          i='hh', j='gender', sep = '_', suffix=r'\w+').reset_index())
cda_frac
```

```python
tts_od = pd.read_csv('TTS OD/tts_od_hh_cleaned.csv')
tts_od
```

```python
df_list = []
for group in ['lep', 'rac', 'ab', 'bl', 'imm']:
    
    temp = tts_od.merge(cda_frac[['hh', 'gender', group]]).copy()
    temp['total'] = temp[group] * temp['total']
    temp = temp.groupby(['o_node', 'd_node', 'period']).sum()['total'].reset_index()
    temp['type'] = group + '_' + temp['period'].str.lower()
    
    temp['origin'] = temp['o_node']
    temp['destination'] = temp['d_node']
    
    temp = temp[['origin', 'destination', 'total', 'type']]
    
    df_list.append(temp)
```

```python
tts_od_type = pd.read_csv('TTS OD/LICO_TTS.csv')
```

```python
tts_od_type = tts_od_type.append(pd.concat(df_list))
```

```python
tts_od_type.to_csv('TTS OD/TTS_TYPE.csv', index = False)
```
