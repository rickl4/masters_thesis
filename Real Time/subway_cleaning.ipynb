```python
import pandas as pd
import datetime
import numpy as np
```

```python
requests_nov = pd.read_csv('2019-11/requests.csv')
response_nov = pd.read_csv('2019-11/responses.csv')
```

```python
response_nov = response_nov.merge(requests_nov)
```

```python
requests_dec = pd.read_csv('2019-12/requests.csv')
response_dec = pd.read_csv('2019-12/responses.csv')
```

```python
response_dec = response_dec.merge(requests_dec)
```

```python
requests_jan = pd.read_csv('2020-01/requests.csv')
response_jan = pd.read_csv('2020-01/responses.csv')
```

```python
response_jan = response_jan.merge(requests_jan)
```

```python
response = response_nov.append(response_dec).append(response_jan)
```

```python
response['date'] = pd.to_datetime(response['create_date']).dt.date
```

```python
response = response[response['date'] > datetime.datetime(2019,11,27).date()].copy()
response = response[response['date'] < datetime.datetime(2020,1,4).date()].copy()
```

```python
response['request_date'] = pd.to_datetime(response['request_date'])
```

```python
response = response.drop(columns = ['system_message_type', 'train_dest', 'data_', 'all_stations'])
```

```python
response
```

```python
response['hour'] = response['request_date'].dt.hour
```

```python
response['date'] = np.where(response['hour'].isin([0,1,2]), (response['request_date'].dt.date - 
                                                   datetime.timedelta(days = 1)).astype(str), 
                       response['request_date'].dt.date.astype(str))
```

```python
response['dow'] = pd.to_datetime(response['date']).dt.weekday
```

```python
response = response[~response['date'].isin(['2019-12-24', '2019-12-25', '2019-12-26', '2019-12-31', '2020-01-01'])]
```

```python
response = response[response['dow'] < 5]
```

```python
date_list = list(response['date'].drop_duplicates())
```

```python
date_list
```

```python
for date in date_list:
    temp = response[(response['date'] == date) & (response['hour'].isin([7,8,9,10,11]))].copy()
    temp['period'] = 'AM'
    temp.to_csv('cleaned_subway/raw_subway_' + date + '_AM.csv', index = False)
    del temp
```

```python
for date in date_list:
    temp = response[(response['date'] == date) & (response['hour'].isin([4,5,6,7,8]))].copy()
    temp['period'] = 'EM'
    temp.to_csv('cleaned_subway/raw_subway_' + date + '_EM.csv', index = False)
    del temp
```

```python
for date in date_list:
    temp = response[(response['date'] == date) & (response['hour'].isin([11,12,13,14,15]))].copy()
    temp['period'] = 'MD'
    temp.to_csv('cleaned_subway/raw_subway_' + date + '_MD.csv', index = False)
    del temp
```

```python
for date in date_list:
    temp = response[(response['date'] == date) & (response['hour'].isin([16,17,18,19,20]))].copy()
    temp['period'] = 'PM'
    temp.to_csv('cleaned_subway/raw_subway_' + date + '_PM.csv', index = False)
    del temp
```

```python
for date in date_list:
    temp = response[(response['date'] == date) & (response['hour'].isin([20,21,22,23,0]))].copy()
    temp['period'] = 'EV'
    temp.to_csv('cleaned_subway/raw_subway_' + date + '_EV.csv', index = False)
    del temp
```
