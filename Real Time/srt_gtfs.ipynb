```python
import pandas as pd
import numpy as np
```

```python
trips = pd.read_csv('gtfs_2019-11-14/trips.txt')
stops = pd.read_csv('gtfs_2019-11-14/stops.txt')
routes = pd.read_csv('gtfs_2019-11-14/routes.txt')
stop_times = pd.read_csv('gtfs_2019-11-14/stop_times.txt')
```

```python
trips = trips[trips['service_id'] == 1]
```

```python
df = trips.merge(routes[routes['route_short_name'] == 3])[['trip_id', 'direction_id']]
```

Kennedy direction is 0.

McCowan direction is 1.

```python
trip_id_list = list(df['trip_id'])
```

```python
stop_times = stop_times[stop_times['trip_id'].isin(trip_id_list)]
stop_times
```

```python
stops[stops['stop_id'].isin(list(stop_times['stop_id'].drop_duplicates()))]
```

```python
stop_times = stop_times[['trip_id', 'arrival_time', 'departure_time', 'stop_id', 'stop_sequence']]
```

```python
int_stop = pd.read_csv('/Volumes/Data2/RST/notebook/GTFS/stop_snapping/GIS Simplify/int_stop.csv')
```

```python
stop_times = stop_times.merge(int_stop[['INT_ID', 'stop_id']]).sort_values(by = ['trip_id', 'stop_sequence'])
```

```python
stop_times['hour'] = stop_times['arrival_time'].str.split(':', expand = True)[0].astype(int)
stop_times['minute'] = stop_times['arrival_time'].str.split(':', expand = True)[1].astype(int)
stop_times['seconds'] = stop_times['arrival_time'].str.split(':', expand = True)[2].astype(int)
```

```python
stop_times = stop_times.drop(columns = ['arrival_time', 'departure_time'])
```

```python
stop_times
```

```python
stop_times['hour'] = np.where(stop_times['hour'] > 23, stop_times['hour'] - 24, stop_times['hour'])
```
