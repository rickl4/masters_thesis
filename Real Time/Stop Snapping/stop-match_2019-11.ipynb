```python
import pandas as pd
import haversine
```

```python
root = '/Volumes/Data2/RST/notebook/'
```

```python
gtfs_new = pd.read_csv(root + 'Real Time/gtfs_2019-11-14/stops.txt')[['stop_id', 'stop_code', 
                                                                      'stop_name', 'stop_lat', 'stop_lon']]

int_stop = pd.read_csv(root + 'GTFS/stop_snapping/GIS Simplify/int_stop.csv')
```

```python
gtfs_new
```

```python
int_stop
```

```python
stop_matching = int_stop.merge(gtfs_new, how = 'outer', on = ['stop_id'], suffixes = ['_16', '_19'])
stop_matching[stop_matching['INT_ID']>1000]
```

```python
stop_matching.info()
```

```python
def hs(df):
    return haversine.haversine((df['stop_lon_16'], df['stop_lat_16']), (df['stop_lon_19'], df['stop_lat_19']),unit=haversine.Unit.METERS)
```

```python
stop_matching['dist'] = stop_matching.apply(hs, axis = 1)
```

```python
stop_matching.sort_values(by = 'dist', ascending = False)
```

```python
unmatched = stop_matching[stop_matching['INT_ID'].isna()][['stop_id','stop_name_19', 
                                                           'stop_lon_19', 'stop_lat_19']]
unmatched
```

```python
unmatched.to_csv('unmatched.csv', index = False)
```
