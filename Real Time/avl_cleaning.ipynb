```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv('avl_2019.tsv', sep='\t', names=['id', 'veh', 'route', 'branch', 'trip', 'direction', 'last_seen-gmt', 'long','lat','sched-time', 'stop_id','neartype'])
```

```python
df = df[~(df['neartype'] == "\\N")]
```

```python
df['stop_id'] = df['stop_id'].str.split('_', expand = True)[0].astype(int)
```

```python
df['timestamp'] = pd.to_datetime(df['last_seen-gmt']).dt.tz_localize('GMT').dt.tz_convert('America/Toronto')
```

```python
df['neartype'] = df['neartype'].astype(int)
df['neartype'] = np.where(df['neartype'] == 0, 1.5, df['neartype'])
```

```python
df = df.sort_values(by = ['id', 'trip','stop_id', 'neartype'])
```

```python
df = df.drop_duplicates(subset = ['id', 'trip', 'direction', 'stop_id'], keep = 'first')
```

```python
df = df.sort_values(by = ['id', 'trip', 'timestamp', 'stop_id'])
```

```python
df[df['id'] == 148762615]
```

```python
df['weekday'] = df['timestamp'].dt.weekday
```

```python
df['hour'] = df['timestamp'].dt.hour
```

```python
df['period'] = None
```

```python
em_df = df.copy()
em_df['period'] = np.where((em_df['weekday'] < 5) & (em_df['hour'].isin([4,5,6,7,8])), 'EM', em_df['period'])

em_df = em_df[em_df['period'] == 'EM']
```

```python
am_df = df.copy()
am_df['period'] = np.where((am_df['weekday'] < 5) & (am_df['hour'].isin([7,8,9,10,11])), 'AM', am_df['period'])

am_df = am_df[am_df['period'] == 'AM']
```

```python
md_df = df.copy()
md_df['period'] = np.where((md_df['weekday'] < 5) & (md_df['hour'].isin([11,12,13,14,15])), 'MD', md_df['period'])

md_df = md_df[md_df['period'] == 'MD']
```

```python
pm_df = df.copy()
pm_df['period'] = np.where((pm_df['weekday'] < 5) & (pm_df['hour'].isin([17,18,19,20,21])), 'PM', pm_df['period'])

pm_df = pm_df[pm_df['period'] == 'PM']
```

```python
ev_df = df.copy()

ev_df['period'] = np.where((ev_df['weekday'] < 5) & (ev_df['hour'].isin([20,21,22,23])), 'EV', ev_df['period'])
ev_df['period'] = np.where((ev_df['weekday'].isin([1,2,3,4,5])) & (ev_df['hour'].isin([0])), 'EV', ev_df['period'])

ev_df = ev_df[ev_df['period'] == 'EV']
```

```python
am_df['date'] = am_df['timestamp'].dt.date.astype(str)
em_df['date'] = em_df['timestamp'].dt.date.astype(str)
pm_df['date'] = pm_df['timestamp'].dt.date.astype(str)
md_df['date'] = md_df['timestamp'].dt.date.astype(str)
```

```python
ev_df['date'] = np.where(ev_df['hour'] == 0, (ev_df['timestamp'].dt.date - datetime.timedelta(days = 1)).astype(str), ev_df['timestamp'].dt.date.astype(str))
```

```python
excluded_dates = ['2019-12-24', '2019-12-25', '2019-12-26', '2019-12-31', '2020-01-01']
```

```python
am_df = am_df[~am_df['date'].isin(excluded_dates)]
pm_df = pm_df[~pm_df['date'].isin(excluded_dates)]
md_df = md_df[~md_df['date'].isin(excluded_dates)]
ev_df = ev_df[~ev_df['date'].isin(excluded_dates)]
em_df = em_df[~em_df['date'].isin(excluded_dates)]
```

```python
dates = list(am_df['date'].drop_duplicates())
```

```python
df_list = [em_df, am_df, md_df, pm_df, ev_df]
```

```python
for df_i in df_list:

    period = df_i['period'].drop_duplicates().iloc[0]
    
    for date in dates:
        temp = df_i[df_i['date'] == date]
        temp[['route', 'trip', 'id', 'veh', 'direction', 'stop_id', 'timestamp', 'period', 'date']].to_csv(
            'cleaned_surface/raw_surface_' + date + '_' + period + '_.csv', index = False)
```
