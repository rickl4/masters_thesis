# Filtering for only the most recent data in a request

Since each request has multiple train arrival times.

We're filtering on `station_char` and `requestid`. This might cause an issue with terminus, but we'll deal with it later.

```python
df_newest = df.drop_duplicates(subset = ['requestid', 'station_char'])
```

```python
df_newest[df_newest['station_char'] == 'KIP1'].sort_values(by = 'request_date').head(30)
```

```python
df_newest[['trainid']].drop_duplicates()
```

```python
single_train = df_newest[df_newest['trainid'] == 162]
```

```python
single_train['estimated_arrival'] = single_train['request_date'] + pd.to_timedelta(single_train['timint'], 'm')
```

```python
single_train = single_train.sort_values(by = 'request_date')
#single_train.head(20)
```

## Creating Trip Ids for Each Train

Terminus are not represented in each trip (ie, its not guaranteed Kipling will show up in the trip)

Trip Ids are only used for the purpose of cleaning up the data, not for any analysis/graph building.

```python
single_train['trip'] = ((single_train['traindirection'] != single_train['traindirection'].shift(1))
                       | ((single_train['request_date'] - single_train['request_date'].shift(1)).dt.seconds > 1800)
                       ).cumsum() - 1
```

```python
single_trip = single_train[single_train['trip'] == 1].copy()
```

## Setting Arrival Times

We're selecting the most recent record where the train is in the station. This means that we will not be attempting to find the departure time. 

API runs once per minute, so the error is limited to minute resolution

```python
single_trip = single_trip[single_trip['timint']<2]
```

```python
arrivals = single_trip[['station_char']].drop_duplicates()
```

```python
arrival_max = single_trip.sort_values(by = 'request_date', ascending = False)
arrival_max = arrival_max.drop_duplicates(subset = ['station_char'])[['station_char', 'estimated_arrival']]
```

```python
arrival_max[arrival_max['station_char'] == 'LWW2']
```

```python
arrivals_at_station = single_trip[single_trip['timint']==0]
arrivals_at_station = arrivals_at_station.sort_values(by = 'estimated_arrival')
arrivals_at_station  = arrivals_at_station.drop_duplicates(subset = ['station_char'])[['station_char', 'estimated_arrival']]
```

```python
arrivals = arrivals.merge(arrival_max, how = 'left')
```

```python
arrivals_at_station = arrivals_at_station.rename(columns = {'estimated_arrival':'at_station_time'})
arrivals = arrivals.merge(arrivals_at_station, how = 'left')
arrivals['arrival'] = True
arrivals
```

```python
arrivals['arrival_time'] = np.where(arrivals['at_station_time'].isna(), arrivals['estimated_arrival'], arrivals['at_station_time'],)
```

We're choosing the most recent record of the train sitting at the station (`at_station_time`) because in some cases, the estimated arrival is later (and in some cases) earlier than the time the train reaches the station. If no such time exists, we're using the estimated arrival time.

```python
arrivals = arrivals.sort_values(by = 'arrival_time').reset_index(drop = True)
```

```python
arrivals
```

Merging back with `single_trip` to regain information from the main df.

```python
arrivals = arrivals[['station_char', 'arrival', 'arrival_time']].rename(columns =
                                                                        {'arrival_time':'estimated_arrival'})
```

```python
len(arrivals)
```

```python
single_trip = single_trip.merge(arrivals)[['station_char', 'subwayline', 'timint', 'stationid', 'traindirection', 'trainid', 'estimated_arrival', 'trip']]
```

```python
single_trip = single_trip.reset_index().merge(single_trip.shift(-1).reset_index()[['station_char', 'estimated_arrival', 'index']], 
                                left_on = 'index', right_on = 'index', suffixes = ['_o', '_d'])
single_trip.dropna()
```
