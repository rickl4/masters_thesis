```python
raw = pd.read_csv('cleaned_wifi/raw_wifi_2019-11-28_AM.csv')
#raw = raw.drop(columns = ['platform1_start_time', 'platform2_end_time', 'station1_end_time', 'station2_start_time'])
```

```python
raw.info()
```

```python
raw['platform1_end_time'] = pd.DatetimeIndex(raw['platform1_end_time'], tz = 'America/Toronto')
raw['platform2_start_time'] = pd.DatetimeIndex(raw['platform2_start_time'], tz = 'America/Toronto')
raw = raw.sort_values(by = 'platform1_end_time')
```

```python
raw = raw[~raw['platform2_start_time'].isna()]
```

# For One Station Pair

```python
test_left = raw[(raw['station1'] == 'ossington') & (raw['station2'] == 'dufferin')].copy()
```

```python
test_left = test_left.reset_index(drop = True).reset_index()
test_right = test_left.copy()
test_right['index'] = test_right['index'] - 1

test_left
```

```python
test = test_left.merge(test_right[['index', 'platform1_end_time', 'platform2_start_time']], left_on = ['index'], right_on = ['index'], suffixes = ['_left', '_right'])
```

```python
test['gap'] = (test['platform1_end_time_right'] - test['platform1_end_time_left']).dt.seconds
test
```

```python
test['arrival_gap'] = ((test['platform2_start_time_right'].astype(int) / 10**9) - (test['platform2_start_time_left'].astype(int) / 10**9)).astype(int)
```

```python
test = test[test['arrival_gap'] > -30]
```

```python
test = test.drop(columns = ['platform1_end_time_right', 'platform2_start_time_right', 'index', 'gap', 'arrival_gap'])
test = test.rename(columns = {'platform1_end_time_left':'platform1_end_time' , 'platform2_start_time_left':'platform2_start_time'})
test = test.reset_index(drop = True).reset_index()
```

```python
test_right = test.copy()
test_right['index'] = test_right['index']-1 
```

```python
test = test.merge(test_right[['index', 'platform1_end_time', 'platform2_start_time']], left_on = ['index'], 
           right_on = ['index'], suffixes = ['_left', '_right'])
test['gap'] = (test['platform1_end_time_right'] - test['platform1_end_time_left']).dt.seconds
```

```python
test['train_id'] = (test['gap'] > 45).cumsum()
```

```python
test[test['train_id'] == 4]['platform1_end_time_left'].median()
```

```python
test[test['train_id'] == 4]['platform2_start_time_left'].median()
```

```python
train_times = []

for train_id in list(test['train_id'].drop_duplicates()):
    station1 = 'ossington'
    station2 = 'dufferin'
    depart_time = test[test['train_id'] == train_id]['platform1_end_time_left'].median()
    arrival_time = test[test['train_id'] == train_id]['platform2_start_time_left'].median()
    
    train_times.append([station1, station2, depart_time, arrival_time])
```

```python
pd.DataFrame.from_records(train_times, columns = ['station1', 'station2', 'depart_time', 'arrival_time'])
```
