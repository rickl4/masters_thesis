```python
import pandas as pd
import numpy as np
```

```python
raw = pd.read_csv('cleaned_wifi/raw_wifi_2019-11-28_AM.csv')
```

```python
raw[['station1', 'station2']].drop_duplicates()
```

```python
raw[raw['station']]
```

```python
raw.info()
```

```python
raw['platform1_start_time'] = pd.DatetimeIndex(raw['platform1_start_time'], tz = 'America/Toronto')
raw['platform1_end_time'] = pd.DatetimeIndex(raw['platform1_end_time'], tz = 'America/Toronto')
raw['platform2_start_time'] = pd.DatetimeIndex(raw['platform2_start_time'], tz = 'America/Toronto')
raw['platform2_end_time'] = pd.DatetimeIndex(raw['platform2_end_time'], tz = 'America/Toronto')
```

```python
test = raw[(raw['station1'] == 'ossington') & (raw['station2'] == 'dufferin')].copy()
```

```python
pd.DatetimeIndex(test['platform1_end_time']).floor('T')
```

```python
test['minute_depart'] = pd.DatetimeIndex(test['platform1_end_time']).floor('T')
test['minute_end'] = pd.DatetimeIndex(test['platform2_start_time']).floor('T')
```

```python
times = test.groupby(['minute_depart', 'minute_end']).count().sort_values(by = 'station1', 
                                                                          ascending = False)[['station1']].reset_index()
times
```

```python
depart = times.sort_values(by = ['minute_depart','station1'], ascending = [True, False])
depart
```

```python
test['delta'] = (test['platform2_start_time'] - test['platform1_end_time']).dt.seconds
```

```python
test['delta'].sort_values()
```

```python
test.groupby('delta').count()
```

```python
test[test['delta'] == 978]
```

```python
self_join = test.merge(test[['station1', 'station2', 'platform1_end_time', 'platform2_start_time']], left_on = ['station1', 'station2'], right_on = ['station1', 'station2'])
```

```python
self_join['depart_delta'] = (self_join['platform1_end_time_x'] - self_join['platform1_end_time_y']).dt.seconds
```

```python
self_join = self_join[self_join['depart_delta']!=0]
self_join
```
